##Import deepC.dnnc

```
import deepC.dnnc as dc
```

# Create Tensors

```
t0 = dc.zeros(3,2)
t1 = dc.ones(2,3)
t2 = dc.array([[1,2,3],[4,5,6]])
t3 = dc.array([[10,20,30],[40,50,60]]).asTypeInt()
print(t2)
```

# Perform Pythonic Operations

All python operators are available on tensors.

```
y=t1 + t2
print(y)

print("."*30)

z= t3/y.asTypeInt()
print(z)
```

# Broadcasting and Implicit Conversion

```
a=True
b=dc.array([[1,2,3],[4,5,6]])
y=a+b; # adding bool and floatTensor of different ranks
y
```

# Tensor to other types

1. Numpy array
1. Python List
1. Python Tuple
 

```
import numpy as np
py_list = list(y);   # convert tensor to python list
py_tuple = tuple(y); # convert tensor to python tuple
np_ary = y.numpy()   # convert tensor to numpy array
```

# Looping throught Tensors

```
for num in dc.arange(5):
  print(num)
```

# Neural Network Operators

100s of NN operators are available

```
#x=dc.arange(4,-4)
#x=dc.array([-4,-3,-2,-1,0,1,2,3,4])
#y=dc.sigmoid(x)

x=dc.arange(1256)/100
sinx=dc.sin(x)
cosx=dc.cos(x)
```

# Plot Data
DNNC tensors support all popular libraries inlcuding matplot, seaborn etc.

## Line Plot

```
import matplotlib.pyplot as plt
plt.plot (x.data(),sinx.data())
plt.plot(x.data(), cosx.data())
```

## Histogram

```
plt.hist(list(dc.array(10000)), bins=50)
plt.show()
```

# Scatter Plot

```
plt.scatter(dc.array(256).data(), 
            list(dc.sin(dc.array(256))),
            c=np.random.rand(256),
            alpha=0.5)
```
